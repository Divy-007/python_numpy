# Numpy cricket simulator

## T-20 Match

#### impoerting libraries

```python
import numpy as np
import matplotlib.pyplot as plt

```

#### defining basic funclaties of match 

```python
runs=[0,1,2,3,4,6]
run_probs=[0.25,0.2,0.2,0.05,0.25,0.05]
wicket_prob=0.08
total_runs=0
total_wickets=0
overs=20
balls=overs*6
```

#### Starting 20 Over innings........ (Team A)

```python
for ball in range(1,balls+1):
    if np.random.rand() <wicket_prob:
        total_wickets+=1
        print(f"ball {ball} WICKET!!! Total wickets:{total_wickets}")
        if total_wickets==10:
          print("Inning ended - All Out!")
          break
    else:
        run=np.random.choice(runs,p=run_probs)
        total_runs+=run
        print(f"Ball {ball} : {run} run(s) | Total : {total_runs}/ {total_wickets}")
print(f"\n Final score of team A : {total_runs} /{total_wickets} in {ball//6}.{ball%6} overs")
```

#### Starting 20 Over innings (Team B)

```python
run_probs=[0.23,0.22,0.2,0.05,0.25,0.05]
wicket_prob=0.09
total_runsB=0
total_wicketsB=0
```

```python
target=total_runs+1
for ball in range(1,balls+1):
    if np.random.rand() <wicket_prob:
        total_wicketsB+=1
        print(f"ball {ball} WICKET!!! Total wickets:{total_wickets}")
        if total_wicketsB==10:
          print("Inning ended - All Out!")
          break
    else:
        run=np.random.choice(runs,p=run_probs)
        total_runsB+=run
        print(f"Ball {ball} : {run} run(s) | Total : {total_runsB}/ {total_wicketsB}")
        if total_runsB>target:
            break 
       
   
       
print(f"\n Final score of team B : {total_runsB} /{total_wicketsB} in {ball//6}.{ball%6} overs")
print(f"\n Final score of team A : {total_runs} /{total_wickets} in {ball//6}.{ball%6} overs")
if(total_runs>total_runsB):
    print(f"team a won the match!!!! by {total_runs-total_runsB} runs")
elif(total_runs<total_runsB):
    print(f"team b won the match!!! by {10-total_wicketsB} wickets ")
else:
    print("match tied!!")

```
